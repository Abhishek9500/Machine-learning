```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('titanic.csv')
```

```python
df.head()
```

#### Note :- In this data there is three mixed data column cabin,ticket and number so split all mixed data 

```python
df['number'].unique()
```

```python
fig=df['number'].value_counts().plot(kind='bar')
fig.set_title("Passengers travlling with")
```

## Work on Number Column

```python
## Extract Numerical part
df['number_numerical']=pd.to_numeric(df['number'],errors='coerce',downcast='integer')
```

```python
# Extract Categorical part
df['number_categorical']=np.where(df['number_numerical'].isnull(),df['number'],np.nan)

df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Ticket'].unique()
```

## Work on Cabin Column

```python
df['cabin_num']=df['Cabin'].str.extract(r'(\d+)')  # Capture numberical part
df['cabin_cat']=df['Cabin'].str[0]  # Capture the first letter
```

```python
df.head()
```

```python
df['cabin_cat'].value_counts().plot(kind='bar')
```

## Work on Ticket column

```python
# extract the last bit of ticket as number
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],
                                   errors='coerce',
                                   downcast='integer')

# extract the first part of ticket as category
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan,
                              df['ticket_cat'])

df.head(20)
```
