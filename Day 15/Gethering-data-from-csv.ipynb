## 2  Opening Csv file From Url

```python
import requests
import pandas as pd
from io import StringIO

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"

headers = {
    "User-Agent": "Mozilla/5.0",
    "Accept": "text/csv"
}

response = requests.get(url, headers=headers)

data=StringIO(response.text)

pd.read_csv(data)
```

 ## 3  Index_col Parameter

index_col="ID" → Uses the column named ID as the index.
index_col=0 → Uses the first column as the index.

```python
df=pd.read_csv('InsuranceData.csv',index_col='PolicyNumber')
df
```

```python
df=pd.read_csv('InsuranceData.csv',index_col=0)
df
```

## 4 Header Parameter

The header parameter in pd.read_csv() specifies which row contains the column names.
1 Default: header=0
Pandas assumes the first row contains column names.
2 header=None
Use this when the file does not have column names.
3. header=1
Use the second row as column names.
4. Custom Column Names
df = pd.read_csv(
    "data.csv",
    header=None,
    names=["Student_ID", "Student_Name", "Student_Age"]
)

## 5 use_cols Parameter

The usecols parameter is used in pd.read_csv() to load only specific columns from a CSV file.

```python
df=pd.read_csv('InsuranceData.csv',usecols=["CustomerID","Gender","Age","ClaimAmount"],index_col=0)
df
```
