# 1. Categorical Data

### (a)  CountPlot

```python
sns.countplot(df['Sex'])
```

### (b) PieChart

```python
counts=df['Sex'].value_counts()
```

```python
counts
```

```python
plt.pie(counts,labels=counts.index,autopct='%1.1f%%')
```

### (b) BoxPlot

```python
sns.boxplot(df['Age'].dropna())
```

### (C) Density plot

```python
sns.kdeplot(df['Age'])
```
